```python
from graphviz import Graph
# Create Digraph object
dot = Graph()
```

```python
# Add nodes 1 and 2
dot.attr(label="(x1=4,5 , x2=3) Z=10,5", labelloc = "t")
dot.attr('node', shape='circle')
dot.node('P0') 
dot.attr('node', shape='doublecircle')
dot.node('P1', xlabel = "\n(x1=4 , x2=2) Z=8")
dot.attr('node', shape='circle')
dot.node('P2', xlabel = "\n(x1=5 , x2=2,4) Z=9,8")
dot.node('P3', xlabel = "\n(x1=5,33 , x2=2) Z=9,33")
dot.attr('node', shape='square')
dot.node('P4', xlabel = "\ninfeasible            ")

# Add edge between 1 and 2
dot.edge('P0', 'P1', xlabel = " x1 ≤ 4 " )
dot.edge('P0', 'P2', label = " x1 ≥ 5 ")
dot.edge('P2', 'P3', xlabel = " x2 ≤ 2 ")
dot.edge('P2', 'P4', label = " x2 ≥ 3 ")

```
